### <font color='blue'>- check unique values in variable</font>

```python
# Check unique values for each column
def unique_value():
    # Create a list of dictionaries for each column's info
    unique_data = []
    
    for col in data.columns:
        unique_values = data[col].unique()
        missing_count = data[col].isna().sum()
        
        unique_data.append({
            'Column': col,
            'Unique_Count': len(unique_values),
            'Missing_Values': missing_count,
            'Unique_Values': unique_values,
            'Data_Type': data[col].dtype
        })
    
    # Convert to DataFrame and return specified columns
    return pd.DataFrame(unique_data)[['Column', 'Unique_Count', 'Missing_Values', 'Data_Type', 'Unique_Values']]

result = unique_value()
result
```

### <font color='blue'>- use histgram to understand each variable</font>

```python
# 1. Plot histograms for numeric columns
data.select_dtypes(include='number').hist(figsize=(20, 15), bins=30)
plt.suptitle("Histograms of Numeric Features", fontsize=16)
plt.tight_layout()
plt.show()
```

```python
# 2. Plot bar charts for categorical columns
categorical_cols = data.select_dtypes(include='object').columns

for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    data[col].value_counts().plot(kind='bar')
    plt.title(f'Bar chart of {col}')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.tight_layout()
    plt.show()
```

## <font color='red'>Step 4: data cleaning/wrangling</font>

### <font color='blue'>- check for missing values</font>

```python
# Check for sum of missing values for each columns
data.isnull().sum()
```

### <font color='blue'>- remove duplicate rows</font>

```python
# Check all duplicates including the first occurrence
duplicates_all = data[data.duplicated(keep=False)]
print("Duplicate Rows (including all occurrences):", len(duplicates_all))
print(duplicates_all.head())

# Show only the duplicates (excluding the first occurrence of each group)
duplicates = data[data.duplicated(keep='first')]
print("Duplicate Rows (excluding first occurrence):", len(duplicates))
print(duplicates.head())
```

```python
# Remove duplicates and keep only the first occurrence
data = data.drop_duplicates(keep='first')

# Confirm removal
print("Data shape after removing duplicates:", data.shape)
```

### <font color='blue'>- handle data quality issues</font>

### ensure that engine size is not 0

```python
# Convert engineSize to numeric
data['engineSize'] = pd.to_numeric(data['engineSize'], errors='coerce')

# ---- BEFORE CLEANING ----
print("Before cleaning (values  0):")
print(data[data['engineSize'] <= 0][['engineSize']])

# Replace invalid engineSize ( 0) with NaN
data.loc[data['engineSize'] <= 0, 'engineSize'] = np.nan

# Fill NaN with median
data['engineSize'].fillna(data['engineSize'].median(), inplace=True)

# ---- AFTER CLEANING ----
print("\nAfter cleaning (values  0):")
print(data[data['engineSize'] <= 0][['engineSize']])
```

### <font color='blue'>- rename columns </font>

```python
# Rename columns to snake_case for clarity
data = data.rename(columns={
    'year':              'manufactured_year',
    'price':             'selling_price',
    'transmission':      'transmission_type',
    'fuelType':          'fuel_type',
    'mpg':               'miles_per_gallon',
    'engineSize':        'engine_size',
})

data.info()
```

### <font color='blue'>final check</font>

```python
#final checking for missing values and data type
print("Missing values per column:")
print(data.isna().sum())

print("\nData types:")
print(data.dtypes)

print("\n")
data.info()
```

## <font color='red'>Step 5: feature engineeering</font>

### <font color='blue'>- derive new data</font>

```python
# car age
data['car_age'] = 2025 - data['manufactured_year']
```

```python
# miles in litre derived from mpg
data['miles_per_litre'] = (data['miles_per_gallon'] / 4.546).round(2)

# Drop the redundant mpg column
data.drop(columns=['miles_per_gallon'], inplace=True)

data.head()
```
